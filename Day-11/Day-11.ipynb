```python
vid = cv2.VideoCapture(0)
while True:
    flag, img = vid.read()
    if flag:
        # All Processing Code Come here

        # plt.imshow(img[:,:,::-1])

        cv2.imshow('Prevview', img)
        key = cv2.waitKey(1)
        if key == ord('q'): # ord is use for find ASCII value 
            break
    else:
        break # if it not reading any input by the camera than break it 
vid.release() # release is use for camera ko release karne ke liye 
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(0)
while True:
    flag, img = vid.read()
    if flag:
        # All Processing Code Come here

        ## To create image into gray scale
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # print(type(img_gray)) #<class 'numpy.ndarray'>
        # break
        th, img_bw = cv2.threshold(img_gray, 100, 200, cv2.THRESH_BINARY)

        # if we use any function than it is in camelCase
        # if we use any constant than it is in UPPERCASE
        # if we use any class than it is in 


        # cv2.imshow('Preview', img_gray)
        cv2.imshow('Preview', img_bw)
        key = cv2.waitKey(1)
        if key == ord('q'): # ord is use for find ASCII value 
            break
    else:
        print('No Frames')
        break # if it not reading any input by the camera than break it 
vid.release() # release is use for camera ko release karne ke liye 
cv2.destroyAllWindows()
```

```python
# To run this file we have to run 'Debug Python File' 




import cv2
import matplotlib.pyplot as py

vid = cv2.VideoCapture(0)
while True:
    flag, img = vid.read()
    if flag:
        # All Processing Code Come here

        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        th, img_bw = cv2.threshold(img_gray, 100, 200, cv2.THRESH_BINARY)

        x,y,w,h = (250, 200, 400, 400)
        img_cropped = img[y:y+h, x:x+w, :]
 
        # cv2.imshow('Preview', img_gray)
        # cv2.imshow('Preview', img_bw)
        cv2.imshow('Preview', img_cropped)

        key = cv2.waitKey(1)
        if key == ord('q'):
            break
    else:
        print('No Frames')
        break 
vid.release() 
cv2.destroyAllWindows()


```
